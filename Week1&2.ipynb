```python
# Function to convert dropwhile elements and returning integer of the list
def dropwhile_to_integers(drop_list):
    return int("".join(str(x) for x in drop_list))

# Recyling dropwhile from above to return permutation integer
for perm in permutations(range(3),3):
    drop_list = list(dropwhile(lambda x: x==0, perm))
    print(dropwhile_to_integers(drop_list))
```

```python
# Function to create rows
def csv_lines(header, line):
    zip_line = zip_longest(header, line, fillvalue=None)
    line_dict = {x[0]: x[1] for x in zip_line}
    return line_dict

# “Read the first line and use string methods to generate a list of all the column names”
with open("sales_record.csv", "r") as file:
    header = file.readline().replace("\n"," ").split(",")
    for i,j in enumerate(file):
        line = j.replace("\n"," ").split(",")
        print(csv_lines(header,line))
        # Breaking the loop after 6 iteration
        if i > 5:
            break
```
